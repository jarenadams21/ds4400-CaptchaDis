```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

def plot_mfcc(audio_path, num):
	# Load the audio file
	y, sr = librosa.load(audio_path)

	# Extract MFCC features
	mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

	# Plot MFCCs
	plt.figure(figsize=(10, 4))
	librosa.display.specshow(mfccs, x_axis='time', cmap='viridis')
	plt.colorbar(format='%+2.0f dB')
	plt.title(f'MFCC of Example {num} Audio Record')
	plt.xlabel('Time')
	plt.ylabel('MFCC Coefficient')
	plt.show()
```

```python
# Example usage
audio_file_path_0 = './data/01/3_01_42.wav'
audio_file_path_1 = './data/27/6_27_42.wav'
audio_file_path_2 = './data/42/9_42_42.wav'

plot_mfcc(audio_file_path_0, 3)
plot_mfcc(audio_file_path_1, 6)
plot_mfcc(audio_file_path_2, 9)
```
